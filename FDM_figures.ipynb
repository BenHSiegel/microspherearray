```python
import numpy as np

import matplotlib.pyplot as plt

frequency = 0.25  # Frequency in Hz
x = np.linspace(0, 20.01*2*np.pi, 1000)
y = np.sin(frequency * x)

fig, ax = plt.subplots(figsize=[4, 2])
ax.plot(x, y)
ax.axis('off')
ax.set_title(r'$\sin(\omega_1 t)$', fontsize=30)
plt.show()
fig.savefig(r'C:\Users\bensi\Downloads\sine waves\sine_wave1.svg', bbox_inches='tight', pad_inches=0)
plt.close()
```

```python
from scipy.signal import argrelextrema

fig,ax = plt.subplots(figsize=[6, 2])
x = np.linspace(0, 19.01*2*np.pi, 1000)
y = np.cos(frequency/3 * x - 3*np.pi/2)
ax.plot(x[100:], (np.sign(y[100:]+0.4) + 1) / 2)
ax.axis('on')
ax.set_title('FDM Spectral Shape', fontsize=20,fontname='Sitka')
ax.set_xlabel('Frequency', fontsize=16,fontname='Sitka')
ax.set_ylabel('Amplitude', fontsize=16,fontname='Sitka')
ax.set_ylim(bottom=0, top=1.3)
ax.set_xlim(left=0)
# Find indices where y is at its local maximum

max_indices = argrelextrema(y[100:], np.greater)[0] + 100  # adjust for slicing
xticks = x[max_indices]
ax.set_xticks(xticks)
ax.set_xticklabels(['$f_1$', '$f_2$', '$f_3$'],fontsize=14)
ax.set_yticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
fig.savefig(r'C:\Users\bensi\Downloads\sine waves\FDMshape1.svg', bbox_inches='tight', pad_inches=0)
```
