```python
import cv2
import os

# Input and output video paths
input_video = r"C:\Users\Ben\Downloads\organize_ending.mp4"  # Replace with your video file path
output_video = r"C:\Users\Ben\Downloads\organize_ending_5x_faster.mp4"

# Open the input video
cap = cv2.VideoCapture(input_video)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Define the codec and create VideoWriter
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

# Process video - keep every 5th frame
frame_count = 0
kept_frames = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Keep only every 5th frame
    if frame_count % 3 == 0:
        out.write(frame)
        kept_frames += 1
    
    frame_count += 1



# Release resources
cap.release()
out.release()

print(f"Processing complete!")
print(f"Original frames: {total_frames}")
print(f"Kept frames: {kept_frames}")
print(f"Output saved to: {output_video}")

```
